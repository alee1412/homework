```python
grouped_price = bed_3_df.groupby('City')['Median Price'].mean()
reset = grouped_price.reset_index()

crime_q1 = np.percentile(median_crime_df['Median'], 25)
crime_q3 = np.percentile(median_crime_df['Median'], 75)

print(crime_q1)
print(crime_q3)

crime_greater = median_crime_df.loc[median_crime_df['Median'] < 4500.25, ['City', 'Median']]
#crime_lesser = crime_greater.loc[crime_greater['Median'] > 1396.625, ['City', 'Median']]

price_crime = pd.merge(reset, crime_greater, on='City', how='inner')

price_crime.rename(columns={'Median': 'Median Crime'},inplace=True)

median_crime = price_crime[['Median Crime']]
median_price = price_crime[['Median Price']]
```

```python
plt.figure(figsize = (20,10))

plt.scatter(median_crime, median_price, s=120, marker='o', color='red', edgecolor='black')
```
